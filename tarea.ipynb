```r
install.packages("googledrive")
install.packages("readr")
library(googledrive)
library(readr)
# Autenticación (esto te pedirá autorización la primera vez)
#drive_auth()

# ID del archivo en Google Drive (reemplaza con el ID real)
datasets_id <- "https://drive.google.com/file/d/1OJTRwYYHcvqz03wqUHjtPHms0xIbvjMv/view?usp=drive_link"

# Crear archivo temporal para guardar el CSV
temp_file <- tempfile(fileext = ".csv")

# Descargar el archivo desde Drive
drive_download(as_id(datasets_id), path = temp_file, overwrite = TRUE)

# Leer el archivo CSV
data <- read_csv(temp_file)

# Ver los primeros datos
head(data)
str(data)
```

Este conjunto de datos describe 14 características de viviendas en Boston, donde:

hay solo 1 variable es cualitativa (categórica):

- CHAS: Variable binaria (0 o 1) que indica si la propiedad limita con el río Charles.

Las 13 restantes son cuantitativas (numéricas):

- Incluyen medidas como tasa de criminalidad (CRIM), número de habitaciones (RM), impuestos (TAX), y el valor mediano de la vivienda (MEDV), entre otras.

```r
# Cargar librería para graficar
library(ggplot2)

# Seleccionar solo columnas numéricas
numeric_data <- data[sapply(data, is.numeric)]

# Transformar en formato largo para graficar con ggplot
library(tidyr)
long_data <- pivot_longer(numeric_data, cols = c(CRIM, MEDV, RM, LSTAT), names_to = "Variable", values_to = "Valor")

# Crear boxplots
ggplot(long_data, aes(x = Variable, y = Valor)) +
  geom_boxplot(fill = "skyblue", color = "darkblue") +
  labs(title = "Boxplots de Variables Cuantitativas", x = "Variable", y = "Valor") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
data_clean <- data[, 1:14]

# Convertimos a formato largo
long_data <- pivot_longer(data_clean, cols = c(ZN, INDUS, NOX, TAX), names_to = "Variable", values_to = "Valor")

# Histograma para cada variable
ggplot(long_data, aes(x = Valor)) +
  geom_histogram(bins = 30, fill = "steelblue", color = "white") +
  facet_wrap(~ Variable, scales = "free", ncol = 4) +
  labs(title = "Histogramas de Variables Cuantitativas del Dataset Boston Housing",
       x = "Valor",
       y = "Frecuencia") +
  theme_minimal()
```

```r
install.packages("e1071")
library(ggplot2)
library(e1071)   # Para skewness y kurtosis
library(dplyr)
# Lista de variables a analizar
vars <- c("CRIM", "TAX", "RM")

# Calcular skewness y kurtosis
data_clean %>%
  select(all_of(vars)) %>%
  summarise(across(everything(),
                   list(skewness = ~skewness(.),
                        kurtosis = ~kurtosis(.)))) %>%
  round(2)
library(tidyr)

long_data <- pivot_longer(data_clean,
                          cols = all_of(vars),
                          names_to = "Variable",
                          values_to = "Valor")

# Graficar densidad
ggplot(long_data, aes(x = Valor, fill = Variable)) +
  geom_density(alpha = 0.5) +
  facet_wrap(~ Variable, scales = "free") +
  labs(title = "Gráficos de Densidad para 3 Variables",
       x = "Valor", y = "Densidad") +
  theme_minimal()
```

```r
# Calcular promedios agrupados por CHAS ----------------- numero 9 ---------
data_chas <- data_clean %>%
  group_by(CHAS) %>%
  summarise(
    Promedio_MEDV = mean(MEDV, na.rm = TRUE),
    Promedio_RM = mean(RM, na.rm = TRUE)
  ) %>%
  pivot_longer(cols = -CHAS, names_to = "Variable", values_to = "Promedio")

# Graficar
ggplot(data_chas, aes(x = factor(CHAS), y = Promedio, fill = Variable)) +
  geom_col(position = "dodge") +
  labs(title = "Comparación de Promedios según cercanía al río Charles (CHAS)",
       x = "¿Cerca del río Charles? (0 = No, 1 = Sí)",
       y = "Promedio",
       fill = "Variable") +
  theme_minimal()
```

```r
# Datos para casas NO cercanas al río ----------------numero 10-----------
grupo_no_rio <- data_clean %>% filter(CHAS == 0)

# Datos para casas SÍ cercanas al río
grupo_si_rio <- data_clean %>% filter(CHAS == 1)

# Verificar tamaños de cada grupo
nrow(grupo_no_rio)
nrow(grupo_si_rio)
```

```r
# ---------------numero 11--------------
vars_cuantitativas <- data_clean %>%
  select(where(is.numeric)) %>%
  select(-CHAS) %>%
  names()

# Calcular media y varianza por grupo
resumen_chas <- data_clean %>%
  group_by(CHAS) %>%
  summarise(across(all_of(vars_cuantitativas),
                   list(media = ~mean(.),
                        varianza = ~var(.)),
                   .names = "{.col}_{.fn}"))

# Mostrar resumen
resumen_chas
```

```r
# ----------------numero 12-----------
vars_cuantitativas <- data_clean %>%
  select(where(is.numeric)) %>%
  names()

# Calcular media y varianza para todo el dataset
resumen_total <- data_clean %>%
  summarise(across(all_of(vars_cuantitativas),
                   list(media = ~mean(.), varianza = ~var(.)),
                   .names = "{.col}_{.fn}"))

# Mostrar el resumen total
resumen_total
```
