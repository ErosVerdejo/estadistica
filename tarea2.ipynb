```r
# Item II "Lanzamiento de una moneda" inciso 5.
set.seed(777696662)
simulacion_1 = rbinom(1000, size = 100, prob = p_cara)
media = 100 * 0.5
ds = sqrt(100 * 0.5 * 0.5)
ggplot(data.frame(x = simulacion_1), aes(x)) + stat_function(fun = dnorm, args = list(mean = media, sd = ds), color = "red", linewidth = 2) + geom_histogram(aes(y = ..density..), binwidth = 1.5, fill = "orange", color = "yellow")
```

```r
# Item II "Lanzamiento de una moneda" inciso 6.
media = 100 * 0.5
ds = sqrt(100 * 0.5 * 0.5)
valor_corregido = 50.5
probNormal_corregida = pnorm(valor_corregido, mean = media, sd = ds)

cat("La probabilidad aproximada usando la Normal con corrección de continuidad es P(X <= 50.5) =", probNormal_corregida, "\n")

probBinomial_exacta = pbinom(q = 50, size = 100, prob = 0.5)
cat("--- Comparación de Resultados ---\n")
cat("Valor exacto (Binomial):", probBinomial_exacta, "\n")
cat("Valor aproximado (Normal con corrección):", probNormal_corregida, "\n")
cat("Diferencia:", probNormal_corregida - probBinomial_exacta, "\n\n")
cat("Al restamos tenemos que la diferencia es de 0.00000000502 lo cual nos demuestra que estos dos metodos nos entregan una precision muy fina en los resultados\n")
cat("pero aún asi el que entrega el valor mas preciso es la binomial\n")
```

**distribución geometrica**

1.- Una variable aleatoria geométrica cuenta el número de intentos (ensayos) hasta obtener el primer éxito en una secuencia de pruebas independientes, donde cada prueba tiene la misma probabilidad de éxito, representada por 'p'. La variable puede tomar valores enteros desde 1 en adelante (1, 2, 3, ...).

: probabilidad de éxito en cada ensayo.


2.- ejemplo: lanzamos una moneda cargada, donde la probabilidad que salga cara es de 0.6

la variable X∼Geom(p=0.6) representa la cantidad de veces que se necesita lanzar una moneda para que salga cara.
la distribución geometrica se usa para fenomenos como :

. numero de veces hasta que un proceso falla o es exitoso.

. numero de personas hasta que alguno compre.

. investigación de mercado, prueba de productos.

. investigación de enfermedades.


```r
#Item IV "Distribución geometrica" inciso 3.
set.seed(732173217)
p1 = 0.6
simulacion_p1 = rgeom(n = 100, prob = p1)

ggplot(data.frame(x = simulacion_p1), aes(x)) + geom_histogram(binwidth = 1.5, fill = "seagreen3", color = "black") +labs(title = "Histograma para 100 valores, parametro = 0.6",x = "", y = "")
```
